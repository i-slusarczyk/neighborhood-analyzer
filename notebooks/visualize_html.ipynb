```python
import geopandas as gpd
import folium
import src.config as cfg
```

```python
gdf = gpd.read_parquet(cfg.PROCESSED_DIR / "h3.parquet")
```

```python
lat = 50.0614721
lon = 19.9379908

m_base = folium.Map(location=[lat, lon], zoom_start=12, tiles=None)

folium.TileLayer(
    tiles="CartoDB Positron",
    name="CartoDB Positron",
).add_to(m_base)
```

```python
m_macro = gdf.assign(
    final_score=gdf["final_score"].round(2),
    median_price=gdf["median_price"].round(0),
    value_ratio=(gdf["value_ratio"] * 15).round(2),
)
```

```python
final_score_map = m_macro.explore(
    column="final_score",
    cmap="RdYlGn",
    tooltip=["final_score", "median_price", "value_ratio"],
    m=m_base,
    name="Final Score",
    show=False,
)
gdf.sort_values(by="final_score").head()
```

```python
value_ratio_map = m_macro.explore(
    column="value_ratio",
    cmap="RdYlGn",
    tooltip=["final_score", "median_price", "value_ratio"],
    m=m_base,
    name="Value Ratio",
    show=True,
)
gdf.sort_values(by="value_ratio").head()
```

```python
median_price_map = m_macro.explore(
    column="median_price",
    cmap="RdYlGn",
    tooltip=["final_score", "median_price", "value_ratio"],
    m=m_base,
    name="Median price",
    show=False,
)
gdf.sort_values(by="median_price").head()
```

```python
folium.LayerControl(collapsed=False).add_to(m_base)
```

```python
m_base.save("krakow_qol_map.html")
```
